```python
def maxMeetings(meetings):
    meetings.sort(key=lambda x: x[1])

    count = 1
    last_end = meetings[0][1]

    for i in range(1, len(meetings)):
        if meetings[i][0] >= last_end:   # FIX HERE
            count += 1
            last_end = meetings[i][1]

    return count


T = int(input())
for _ in range(T):
    N = int(input())
    meetings = []
    for _ in range(N):
        s, e = map(int, input().split())
        meetings.append((s, e))

    print(maxMeetings(meetings))
```

```python
def first_occurrence(arr, x):
    low, high = 0, len(arr) - 1
    ans = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            ans = mid
            high = mid - 1
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return ans


def last_occurrence(arr, x):
    low, high = 0, len(arr) - 1
    ans = -1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            ans = mid
            low = mid + 1
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1
    return ans


T = int(input())
for _ in range(T):
    N, Q = map(int, input().split())
    arr = list(map(int, input().split()))

    query_values = list(map(int, input().split())) # Read all Q query values from a single line
    for x in query_values: # Iterate through each query value
        first = first_occurrence(arr, x)
        if first == -1:
            print(0)
        else:
            last = last_occurrence(arr, x)
            print(last - first + 1)
```
